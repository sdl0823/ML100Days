### 1. [簡答題] 請問下列四種不同的 DataFrame 選取結果有什麼差異？

```
df.loc[ '2013-01-01', 'A'] 
df.loc[ '2013-01-01', ['A', 'B'] ] 
df.loc[ '2013-01-01':'2013-01-02', 'A' ] 
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] 
```





````python
import pandas as pd
import numpy as np
dates = pd.date_range('1/1/2013', periods=8)
df = pd.DataFrame(np.random.randn(8, 4),
                  index=dates, columns=['A', 'B', 'C', 'D'])
df
````

````python
print(f"df.loc['2013-01-01', 'A']值: {df.loc['2013-01-01', 'A']},\t type: {type(df.loc['2013-01-01', 'A'])}")
print(f"df.loc['2013-01-01', ['A', 'B']]值:\n{df.loc['2013-01-01', ['A', 'B']]},\t \
      type: {type(df.loc['2013-01-01', ['A', 'B']])}")
print(f"df.loc['2013-01-01':'2013-01-02', 'A']值:\n{df.loc['2013-01-01':'2013-01-02', 'A']},\t \
      type: {type(df.loc['2013-01-01':'2013-01-02', 'A'])}")
print(f"df.loc['2013-01-01':'2013-01-05', 'A':'C']值:\n{df.loc['2013-01-01':'2013-01-05', 'A':'C']},\t \
      type: {type(df.loc['2013-01-01':'2013-01-05', 'A':'C'])}")
````

```

# - select the last 2 columns.
# - select the even columns. (index = 0, 2, 4)
```


````python
# 記得先 Import 正確的套件
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
````

````python
# - select the first 3 rows.
df[:3]
````

````python
# - select the odd rows. (index = 1, 3, 5)
df[1::2]
````

````python
# - select the last 2 columns.
df.iloc[:,[0,2]]
````

````python
# - select the even columns. (index = 0, 2, 4)
df.iloc[:,::2]
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
````

````python
# - 1. filtered by first column > 20?
df[df[0] > 20]
````

````python
# - 2. filtered by first column + second column > 50
df[df[0]+df[1] > 50]
````

````python
# - 3. filtered by first column < 30 or second column > 30
df[(df[0] < 30) | (df[1] > 30)]
````

````python
# - 4. filtered by total sum of row > 100
df[df.sum(axis=1) > 100]
````
